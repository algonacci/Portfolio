### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('dark_background')
```

### Load Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/algonacci/DataWarehouse/main/AirBnb.csv')
df
```

### Exploratory Data Analysis

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

### Data Preprocessing

```python
### Choosing only numeric column
df[['price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365']].describe()
```

```python
df.drop(['name','host_name'],axis=1, inplace=True)
```

```python
### Imputation with modes

modus = df['last_review'].mode()[0]

df['last_review']=df['last_review'].fillna(modus)
```

```python
### Imputation with mean

df['reviews_per_month']=df['reviews_per_month'].fillna(df['reviews_per_month'].mean())
```

```python
df.isnull().sum()
```

#### Adding Price Category

```python
df["price_category"] = np.where(df["price"] < 150, "Cheap", np.where( df["price"] <= 500, "Medium", "Expensive"))
df[["host_id","neighbourhood_group","price_category"]]
```

```python
df['price_category'].value_counts()
```

```python
### Filter data to display data above 2019 only

filtered_date = df.loc[(df['last_review'] >= '2019-01-01')]

filtered_date
```

```python
filtered_date['price_category'].value_counts()
```

### Data Visualization

```python
corr = df.corr(method='kendall')
plt.figure(figsize=(15,8))
plt.title('Data Correlation', fontsize=20, pad=20)
sns.heatmap(corr, annot=True)
```

```python
plt.figure(figsize=(8,8))
sns.countplot(x='price_category', data=filtered_date, palette="plasma")
plt.title('Number of Accommodation Price Categories on AirBnb', fontsize = 20, pad=20)
plt.ylabel('Number of Accommodation', fontsize = 20)
plt.xlabel('Price Category', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='neighbourhood_group', data=filtered_date, palette="plasma")
plt.title('Division of Accommodation Region on AirBnb', fontsize = 20, pad=20)
plt.ylabel('Number of Accommodation', fontsize = 20)
plt.xlabel('Region', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x='room_type', data=filtered_date, palette='plasma')
plt.title('Comparison of Accommodation Types on AirBnb', fontsize = 20, pad=20)
plt.xlabel('Accommodation Types', fontsize = 20)
plt.ylabel('Number of Accommodation', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(df.longitude,df.latitude,hue=df.neighbourhood_group, palette='plasma')
plt.title('Map of Neighbourhood group', pad=20, fontsize=20)
plt.ioff()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(df.longitude,df.latitude,hue=df.room_type)
plt.title('Map of Room Typesx', pad=20, fontsize=20)
plt.ioff()
```
